```python
import os
import pandas as pd
from IPython.display import display
from matplotlib.backends.backend_pdf import PdfPages

path = "D:\\eclipse projects\\ODIN_data\\tex_study_34"

def get_all_files(path: str) -> list:
    files = []
    # r=root, d=directories, f = files
    for r, d, f in os.walk(path):
        for file in f:
            files.append(os.path.join(r, file))
    return files
```

```python
files = get_all_files(path)
dfs = []
for file in files:
    df = pd.read_csv(file)
    df['range'] = file[file.rindex('t_')+2:-4]
    df.rename(columns={'_couponName': "coupon_name"}, inplace = True)
    dfs.append(df)
    display(df)
```

```python
def get_title() -> str:
    return "\\documentclass[landscape]{article} \\usepackage[utf8]{inputenc} \\usepackage{multirow} \
\\usepackage[table]{xcolor} \\usepackage[margin=0.5in]{geometry} \n \\title{Daily Report for Study: XYZ} \
\\author{maisha.jauernig } \\begin{document} \\maketitle \\newpage \n\n"

def get_col(num_cols: int, rids: list) -> str:
    table = "\\begin{tabular}{" 
    for i in range(num_cols):
        table += "|c"
    table += "|} \\hline \\multirow{2}{*}{coupon} "
    
    for i in rids:
        table += "& \\multicolumn{" + str(num_subcols) + "}{c|}{Rule " + str(i) +"}"
    table += "\\\\ \\cline{2-" + str(num_cols) + "} \n"
    
    for i in range(len(rids)):
        table += " & answered/fired & fired/ideal fires & ideal fires "
    table += "\\\\ \\hline \\hline \n"
    
    return table

def get_row(df: pd.DataFrame, df_col_idx: int, num_cols: int) -> str:
    table = ""
    for r in range(len(df)):
        table += df.iloc[r]['coupon_name']
        
        for i in range(df_col_idx, df_col_idx+num_cols, 4):
            ans = df.iloc[r,i]
            fired = df.iloc[r,i+1]
            ideal = df.iloc[r,i+2]
            ans_of_fired = ans/fired if fired > 0 else 0
            fir_of_ideal = fired/ideal if ideal > 0 else 0
            table += " & " + (" \\cellcolor{red!25} " if ans_of_fired < .5 else " ") + str(ans_of_fired) + " & " + \
                    (" \\cellcolor{red!25} " if fir_of_ideal < .5 else " ") + str(fir_of_ideal) + " & " + str(ideal)
                
        table += "\\\\ \\hline \n"
        
    return table

def get_one_table_page(df: pd.DataFrame, all_rids: list, num_subcols: int, df_col_idx: int) -> str:
    date = df['range'][0]
    date = date[date.rindex('to_')+3:]

    table = "\\begin{center} {\\huge " + date + " \\\\ } \n"

    for rid_idx in range(0, len(all_rids), 2):

        rids = all_rids[rid_idx:rid_idx+2]
        num_cols = len(rids) * num_subcols + 1

        table += get_col(num_cols, rids)
        table += get_row(df, df_col_idx, num_cols)
        table += "\\\\ \\end{tabular} \n"

        df_col_idx += num_cols + 1

    table += "\\end{center} \\newpage \n\n"
    
    return table
```

```python
table = get_title()

for df in dfs:
    all_rids = []
    for i in range(1, len(df.columns)-1, 4):
        all_rids.append(df.iloc[0,i])

    num_subcols = 3
    df_col_idx = 2

    date = df['range'][0]
    date = date[date.rindex('to_')+3:]

    table += get_one_table_page(df, all_rids, num_subcols, df_col_idx)

table += "\\end{document}"
print(table)
```
